```python
from __future__ import absolute_import, division, print_function

#計算やデータ処理のライブラリインポート
import pathlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
#データの取得
df = pd.read_csv("../shizuoka/population.csv", encoding="UTF-8")
df
```

```python
#可視化
sns.pairplot(df[["age0to19", "age20to39", "age40to59", "age60to79","age80over","house"]], diag_kind="kde")
```

```python
#houseにNaNが存在するため、前処理
df = df[df['house'] > 0]

# 特徴量とターゲットに切り分け
X_data = np.array(df[['age0to19', 'age20to39', 'age40to59', 'age60to79','age80over']])
y_data = np.array(df[['house']])
 
# １行目のデータの特徴量（X)とターゲット（y）を確認
print(X_data[0:1])
print(y_data[0:1])
```

```python
# 正規化
def norm(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    return (data - mean) / std
 
# データを変更
X_data = norm(X_data)
print(X_data[0:1])
```

```python
# 1を追加する前のサイズ
print(X_data.shape)
 
# 1を作成
ones = np.ones((842, 1))
 
# 1を追加
X_data = np.c_[ones, X_data]
X_data.shape
```

```python
# 訓練データとテストデータへ切り分け　今回テストデータは20%に
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
y_train = y_train.reshape(673,1)
y_test = y_test.reshape(169,1)

#訓練データxy,testデータxyの順
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 学習率とエポック（反復処理回数）
learning_rate = 0.01
training_epochs = 100

# 特徴量の数
n_dim = X_data.shape[1]

# 特徴量（X)とターゲット（y）のプレースホルダー
X = tf.placeholder(tf.float32,[None,n_dim])
Y = tf.placeholder(tf.float32,[None,1])

# 係数（W）と定数項（b）の変数
W = tf.Variable(tf.ones([n_dim,1]))
b = tf.Variable(0.0)
```

```python
# 線形モデル
y = tf.add(b, tf.matmul(X, W))

# コスト関数
cost = tf.reduce_mean(tf.square(y - Y))

# 最適化
training_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
cost_history = [0]
# 初期化
init = tf.global_variables_initializer()
# モデル訓練開始
sess = tf.Session()
sess.run(init)
for epoch in range(training_epochs):
    sess.run(training_step, feed_dict={X:X_train, Y:y_train})
    cost_history = np.append(cost_history, sess.run(cost, feed_dict={X:X_train, Y:y_train}))
    if epoch % 100 == 0:
        W_val = sess.run(W)
        b_val = sess.run(b)
```

```python
# 誤差（cost）を確認
print(cost_history[1])
print(cost_history[50])
print(cost_history[100])
```

```python
# テストデータを使って予測
pred_test = sess.run(y, feed_dict={X: X_test})
```

```python
pred = pd.DataFrame({"実際のhouse":y_test[:,0], "予測したhouse":pred_test[:,0]})
pred.head()
```
