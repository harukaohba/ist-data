## 3. データ可視化： グラフ作成
### 1. データ準備
e-Stat人口推計より、
[都道府県5歳階級別人口データ](https://www.e-stat.go.jp/stat-search/files?page=1&layout=datalist&toukei=00200524&tstat=000000090001&cycle=7&year=20170&month=0&tclass1=000001011679
)をダウンロード

### 2. データのアップロード・グラフ作成準備
データをアップロードし、ラベルカラムを作成

```r
#csvファイルの読み込み
csv_data <- read.csv("prefecture.csv")
csv_data 
```

```r
#日本語の対応
par(family = "HiraKakuProN-W3") 
```

```r
#グラフで用いるラベルを作成
labels=c("0-4歳","5-9歳","10-14歳","15-19歳","20-24歳","25-29歳","30-34歳","35-39歳","40-44歳","45-49歳","50-54歳","55-59歳","60-64歳","65-69歳","70-74歳","75-79歳","80-84歳","85歳以上")
labels
```

### 3.円グラフ
東京都の人口グラフを作成する

```r
index <- 13 #indexは行番号を指定
pie(as.integer(c(csv_data[index,4:21])),labels = labels, 
    main = csv_data[index,2], clockwise=TRUE, col = rainbow(18))
```

### 4.棒グラフ

```r
barplot(as.integer(c(csv_data[index,4:21])),legend = labels,
        main =csv_data[index,2], names.arg = labels, col = rainbow(18))
```

### 5.積み上げグラフ

```r
#都道府県別年齢層
barplot(t(csv_data[,4:21]), legend = labels,
        col = rainbow(18), names.arg = csv_data[,2]) 
```

```r
#ソートしたものを表示する
des_data <- csv_data[order(csv_data$総数, decreasing = TRUE),]
#des_data
barplot(t(des_data[,4:21]), legend = labels,
        col = rainbow(18), names.arg = des_data[,2])
```

### 6.散布図
各都道府県の総数と70歳以上の人数の散布図、回帰直線をかく

```r
#選択する年齢層のリストを作ります。
columnList <- c("総数", "X70.74歳","X75.79歳","X80.84歳","X85歳以上")
#70歳以上の人数の和を求める
scatter_y <- csv_data[,18]+csv_data[,19]+csv_data[,20]+csv_data[,21]
#プロット
plot(csv_data[,1], scatter_y, xlab = "Population", ylab = "70 Over")
#回帰直線を求める
result.lm <- lm(scatter_y ~ csv_data[,1])
#回帰直線の内容を表示
summary(result.lm)
#回帰直線を追加
abline(result.lm)
#R値を追加
mtext(substitute(paste(R^2," = ", text),
                 list(text = round(summary(result.lm)$adj.r.squared,digits = 5))),
      line=-1,col=2,adj=0)
```
