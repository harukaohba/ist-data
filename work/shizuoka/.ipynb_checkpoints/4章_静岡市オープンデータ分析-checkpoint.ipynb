## 4-10. 主成分分析

第２主成分：第１主成分に直交し、ばらつきが２番目に大きい  
寄与率：その成分の分散が占める割合  
  
基準

## 4-11. 平均の差の検定の準備
葵区、駿河区データで世帯数を比較する。  

```r
#世帯数データ作成
preaoi.house <- csv_data[csv_data$区=="葵区","世帯数"]
presuruga.house <- csv_data[csv_data$区=="駿河区","世帯数"]
```

## 4-12. F検定の実施
帰無仮説(H0)：「2群間の分散に差がない(等分散である)」  
対立仮説(H1)：「2群間の分散に差がある(等分散でない)」

```r
#葵区と駿河区の世帯数
var.test(preaoi.house,presuruga.house)
#この場合、p-value=2.2e-16 < 0.05なので帰無仮説が棄却
```

## 4-13. 平均の差の検定 (t検定)
帰無仮説：「二群の母平均は等しい」  
対立仮説：「二群の母平均には差がある」
    
両側検定と片側検定では、両側検定のほうが厳しい条件であるため、片側検定のほうが棄却されやすくなります。(片側の方が甘くなる)  
両側検定にするか片側検定にするかは、実験の目的に合わせて検定を行う前に決めておかなくてはなりません。  
両側検定をやってみて帰無仮説が棄却されないからといって、当初の実験計画から外れて片側検定をやる、というのは誤りです。  
上片側検定：alternative=“greater”  
下片側検定：alternative=“less”  
両側検定：alternative=“two.sided”

```r
t.test(preaoi.house, presuruga.house, var.equal = T)
#この場合、p-value = 1.013e-12 < 0.05 となり、帰無仮説は棄却される。２群に差がある。
```

## 4-14. 3群比較の準備
各区総数の多い方から上位15町を選び出す  
region番号1 -> 葵区  
region番号2 -> 駿河区  
region番号3 -> 清水区

```r
population45.data <- read.csv("population15_3.data.csv")
population45.data 
```

## 4-15. bartlett検定
帰無仮説：「k群間の分散は等しい」  
対立仮説：「k群間の分散には差がある」  
区ごとに0歳から19歳の人口を検定する  
    
bartlettで棄却された場合：スティール・ドゥワス(Steel-Dwass)を用いる  
棄却されない場合：bonferroniを用いる

```r
bartlett.test(population45.data[,2],population45.data[,1])
#この場合、p-value = 0.003 < 0.05 となり、帰無仮説は棄却される。
```

## 4-16. スティール・ドゥワス(Steel-Dwass)検定
帰無仮説：「 k群間は似ている」  
対立仮説：「 k群間は差がある」  
区ごとに0歳から19歳の人口を検定する  

```r
source("http://aoki2.si.gunma-u.ac.jp/R/src/Steel-Dwass.R", encoding="euc-jp")
Steel.Dwass(population45.data[,2], population45.data[,1])
#この場合、p-value < 0.05 であると、帰無仮説は棄却される。
#葵区-駿河区：差がある、駿河区-清水区：差がない
```

## 4-17. bonferroni検定
帰無仮説：「 k群間は似ている」  
対立仮説：「 k群間は差がある」  
区ごとに0歳から19歳の人口を検定する

```r
pairwise.t.test(population45.data[,2],population45.data[,1], p.adjust.method="bonferroni")
#この場合、p-value < 0.05 であると、帰無仮説は棄却される。
```
