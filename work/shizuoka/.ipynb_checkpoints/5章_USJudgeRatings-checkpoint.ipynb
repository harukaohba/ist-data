# USJudgeRatings
アメリカ高等裁判所判事43名の評価データを利用する  
  
CONT : 弁護士が裁判官と接触した回数、INTG : 判決の無欠性、  
DMNR : 態度、DILG : 勤勉、CFMG : 裁判の進行管理、DECI : 迅速な判決、  
PREP : 裁判に備える準備、FAMI : 法律に熟知、  
ORAL : 口頭による適切な判決、WRIT : 書面による適切な判決、  
PHYS : 身体的能力、RTEN : 人物性  
  
基本分析をして、データを把握しなさい。  
クラスタリングを実行して、43名を分類しなさい

## クラスタリング

```r
#階層型手法(凝集型階層)
library(cluster)
cluster.agnes <- agnes(USJudgeRatings, stand = T)
plot(cluster.agnes, which = 2, main = "", xlab = "", ylab = "")
```

```r
#区分型階層(全体から区分していく）
cluster.diana <- diana(USJudgeRatings, stand = T) 
plot(cluster.diana, which = 2, main = "", xlab = "", ylab = "") 
```

```r
#区切りを入力
cluster.h <- hclust(dist(USJudgeRatings), method = "ward.D2")
plot(cluster.h, hang=-1) 
rect.hclust(cluster.h, k = 3, border = "red") 
```

## k-means法

```r
set.seed(123)
cluster.kmeans = kmeans(USJudgeRatings, 5)
cluster.kmeans$cluster
table(cluster.kmeans$cluster)
cluster.kmeans$centers
```

```r
clusplot(USJudgeRatings, cluster.kmeans$cluster, color=TRUE, shade=TRUE, labels=2, lines=0)
```
