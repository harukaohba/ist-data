# 7章アンケート分析
次のようなアンケートを集計して、レポートを作成しなさい。  
総合評価を伸ばすにはどの項目が効果的ですか？  
  
アンケート名：余暇の過ごし方に関するアンケート  
アンケート項目：  
1.所属を教えてください。  
2.年齢層を教えてください。  
3.あなたは宿泊を伴う旅行にどの程度行きますか？(年間)  
4.あなたは海外旅行にどの程度行きますか？(年間)  
5.あなたは飛行機をどの程度利用しますか？（年間）  
6.あなたは本をどの程度読みますか？（月間）  
7.あなたは週にどの程度外食をしますか？(週間)  
8.あなたの余暇の過ごし方の総合評価をしてください(0:悪い – 4:良い)  

## データ前処理
Indexをわかりやすい単語に変更：部署 -> Department  
「回」、「冊」、「以上」をとって数値のみにする  
CSVに変換して、アップロード (ファイル名 freetime.csv)  
Department, Ageは属性として利用  

```r
#csvファイルの読み込み
csv_data <- read.csv("freetime.csv")
csv_data
```

```r
summary(csv_data)
```

## 可視化

```r
par(ps = 7)#文字の大きさを小さく
boxplot(csv_data)
```

```r
par(mfrow=c(4,2)) 
# for (ループ変数 in ベクトルやリスト)
for (i in 2:8) {
    hist(csv_data[,i], xlab =colnames(csv_data[i]), main = colnames(csv_data[i]))
}  
```

```r
library(psych) 
pairs.panels(csv_data[2:8])
```

```r
#主成分分析の実行
pca = prcomp(csv_data[3:8], scale = T)
pca
summary(pca)
biplot(pca)
```

## クロステーブルの作成

```r
xtabs(~ Department + Stay, data = csv_data)
xtabs(~ Department + Foreign, data = csv_data)
xtabs(~ Department + Plane, data = csv_data)
xtabs(~ Department + Book, data = csv_data)
xtabs(~ Department + Eat, data = csv_data)
xtabs(~ Department + Total, data = csv_data)
```

```r
#グラフ作成
color = c("gray", "blue","yellow","green","pink")
legendname = rownames(xtabs(~ Stay + Department, data = csv_data))
par(ps = 8)
barplot(prop.table(xtabs(~ Stay + Department, data = csv_data),margin=2), col=color,legend = legendname, main = "Stay")
barplot(prop.table(xtabs(~ Foreign + Department, data = csv_data),margin=2), col=color,legend = legendname, main = "Foreign")
```

## Stepwise

```r
#Totalを目的変数として、Stepwiseで重回帰分析を行う
library(tidyverse)
library(caret, warn.conflicts=F)
library(leaps)
library(MASS, warn.conflicts=F)

full.model <- lm(Total ~., data = csv_data[,3:8])
step.model <- stepAIC(full.model, direction = "both", trace = TRUE)
summary(step.model)
```
