# 7章　演習問題 : swiss
swissデータセット を利用して、回帰分析を行う。  
  
[swissデータセット ]   
1888年のスイス州ごとの出生率，社会経済的指標データ   
データフレーム　47×6   
Fertility : 普通に標準化された出生率測定値   
Agriculture : 男性の農業従事者率(%)   
Examination : 徴兵検査で最高位にマークされる召集兵の割合   
Education : 召集兵のうち小学校以上の学歴のあるものの割合   
Catholic : カソリック信者の割合   
Infant.Mortality : 1年未満で死亡する乳児の生産児に対する割合  

相関係数行列を求め、相関の有無を確認しなさい  
Fertilityを目的変数、 Agricultureを説明変数として単回帰分析を行いなさい  
Fertilityを目的変数、 AgricultureとExaminationを説明変数として重回帰分析を行いなさい  
Fertilityを目的変数とし、Stepwiseを利用して説明変数を選択し、重回帰分析を行いなさい

```r
par(ps = 10)
boxplot(swiss)
```

```r
library(psych)
pairs.panels(swiss)
#Fertility(出生率)とExamination(徴兵検査成績)とEducation(学歴)は負の相関
#AgricultureとExamination,Educationは負の相関
#CatholicとExaminationが負の相関、Agricultureが正
```

## 単回帰分析
目的変数 : Fertility(出生率)  
説明変数 : Agriculture(農業従事率) 

```r
plot(swiss[,2], swiss[,1])
result.lm <- lm(Fertility ~ Agriculture, data = swiss)
summary(result.lm)
abline(result.lm)
#変数は有意だが、R値がかなり低い
```

## 重回帰分析
目的変数 : Fertility(出生率)  
説明変数 : Agriculture(農業従事率), Examination(徴兵検査成績)  

```r
result.lm <- lm(Fertility ~ Agriculture + Examination, data = swiss)
summary(result.lm)
#R値は上昇説明変数が有意でない
```

## Stepwiseで分析
目的変数 : Fertility(出生率)

```r
library(tidyverse)
library(caret, warn.conflicts=F)
library(leaps)
library(MASS, warn.conflicts=F)

full.model <- lm(Fertility ~., data = swiss)
step.model <- stepAIC(full.model, direction = "both", trace = TRUE)
summary(step.model)

#R値が上昇説明変数も有意になっている
#説明変数の選択が特徴的
```
