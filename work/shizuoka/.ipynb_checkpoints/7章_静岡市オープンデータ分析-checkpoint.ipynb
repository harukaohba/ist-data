## ７章　データの予測
-----

## 7-1. データの取得
全ての町で、年齢層別に人数をまとめてデータを作成  
0歳から19歳、20歳から39歳、40歳から59歳、60歳から79歳、80歳以上、世帯数

```r
population.data <- read.csv("population.csv")
population.data 
```

## 7-2. 単回帰分析
今回はpopulation.dataのhouse(世帯数)とage80over(高齢者人数)を利用する  
まず散布図を描画

```r
plot(population.data$house, population.data$age80over)
```

```r
#回帰直線を求める関数lmを利用
#result.lmに結果を格納
result.lm <- lm(age80over ~ house, data=population.data)
summary(result.lm)
```

```r
plot(population.data$house, population.data$age80over)
#描画 : abline
abline(result.lm)
```

世帯数を指定して、高齢者数を予測する。  
3000世帯の時の、高齢者数を予測する

```r
#予測用データ作成
predict.data <- data.frame(house = c(3000))
#予測
predict <- predict(result.lm, predict.data, interval = "prediction")
#予測値、最小予測値、最大予測値の順で表示
```

## 7-3. 重回帰分析
y = a x1 + b x2 +c の形

```r
result.lm <- lm(age80over ~ house + age0to19, data=population.data)
summary(result.lm)
```

```r
#世帯数3000、未成年数500の時、高齢者数を予測
predict.data <- data.frame(house = 3000, age0to19 = 500)
predict <- predict(result.lm, predict.data, interval = "prediction")
predict
```

## 7-4. 重回帰分析 : Stepwise
高齢者数を求めるときに、どの説明変数を使えば効果的かを検証します。  
パッケージのインストール : Terminalで実施  
install.packages("leaps")


```r
#ライブラリの取り込み
library(tidyverse)
library(caret, warn.conflicts=F)
library(leaps)
library(MASS, warn.conflicts=F)
#実行
full.model <- lm(age80over ~., data = population.data)
step.model <- stepAIC(full.model, direction = "both", trace = TRUE)
summary(step.model)
```
