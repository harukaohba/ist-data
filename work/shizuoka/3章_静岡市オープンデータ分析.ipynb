## データインポート

```r
#csvファイルの読み込み
csv_data <- read.csv("shizuoka_R.csv")
csv_data 
```

## グラフ作成

```r
#日本語の対応
par(family = "HiraKakuProN-W3") 
```

```r
#グラフで用いるラベルを作成
labels=c("0-4歳","5-9歳","10-14歳","15-19歳","20-24歳","25-29歳","30-34歳","35-39歳","40-44歳","45-49歳","50-54歳","55-59歳","60-64歳","65-69歳","70-74歳","75-79歳","80-84歳","85歳以上")
```

### 円グラフ
1行目のデータ(葵区大手町)に対して、グラフを書いていく  
 `as.integer(c(csv_data[index,5:22]) `5列目から22列目までのデータを数値型に  
 `labels = labels ` labelsはIn[3]で作成したデータを用いる  
 `main =  paste(csv_data[index,2],csv_data[index,3]) `グラフタイトルを2列目と3列目のデータとする  
 　　　paste() は文字列ベクトルの結合  
 `clockwise=TRUE` 円グラフのデータのスタートを12時から  
 ` col = rainbow(18)` 虹色で12段階を指定

```r
index <- 1 #indexは行番号を指定
pie(as.integer(c(csv_data[index,5:22])),labels = labels, 
    main = paste(csv_data[index,2],csv_data[index,3]),
    clockwise=TRUE, col = rainbow(18))
```

### 棒グラフ

```r
barplot(as.integer(c(csv_data[index,5:22])),legend = labels,
        main = paste(csv_data[index,2],csv_data[index,3]),
        names.arg = labels, col = rainbow(18))
```

#### 積み上げグラフ
最初の10件の町別年齢層について可視化  
   
 `t() `は転置する関数 : `t(csv_data[1:10,5:22])` データを取得し、転置

```r
barplot(t(csv_data[1:10,5:22]), legend = labels,
        col = rainbow(18), names.arg = csv_data[1:10,3]) 
#10件の町別年齢層
```

```r
#100件の町別年齢層
barplot(t(csv_data[1:100,5:22]), legend = labels,
        col = rainbow(18), names.arg = csv_data[1:100,3])
```

人口の降順に並び替えて表示をする

```r
des_data <- csv_data[csv_data$区=="葵区",] #葵区だけの町を選択
des_data <- des_data[order(des_data$総数, decreasing = TRUE),] #人口で並び替え
des_data #データを表示
```

```r
#ソートしたものを表示する
barplot(t(des_data[1:100,5:22]), legend = labels,
        col = rainbow(18), names.arg = des_data[1:100,3])
```

#### 散布図を作成
各区の総数と70歳以上の人数のグラフを描く

```r
#選択する年齢層のリストを作成
columnList <- c("総数", "X70.74歳","X75.79歳","X80.84歳","X85歳以上")
#葵区の総数と70歳以上の人数を選択
scatter <- csv_data[csv_data$区=="葵区",columnList,]
#70歳以上の人数の和を求める
scatter_y <- scatter[,2]+scatter[,3]+scatter[,4]+scatter[,5]
#プロット
plot(scatter[,1], scatter_y, xlab = "Population", ylab = "70 Over")
```

回帰直線を求める  
今回は、  
`scatter_y(70歳以上人口)` = 0.189785 * `scatter[,1](総数)` + 13.664494  
という回帰式が導出された

```r
#回帰直線を求める
result.lm <- lm(scatter_y ~ scatter[,1])
#回帰直線の内容を表示
summary(result.lm)
```

```r
#プロット
plot(scatter[,1], scatter_y, xlab = "Population", ylab = "70 Over")
#回帰直線を追加
abline(result.lm)
#R値を追加
mtext(substitute(paste(R^2," = ", text),
      list(text = round(summary(result.lm)$adj.r.squared,digits = 5))),
      line=-1,col=2,adj=0)
```
