## ５章　データの分類1：教師なし機械学習クラスタ分析
-----

## 5-1. クラスタ分析：階層型手法(凝集型階層)
凝集型階層（観測値からクラスタへまとめていく）  

```r
library(cluster)
cluster.agnes <- agnes(population15.data, stand = T)
plot(cluster.agnes, which = 2, main = "", xlab = "", ylab = "")
```

## 5-2. クラスタ分析：階層型手法(区分型階層)
区分型階層(全体から区分していく）

```r
cluster.diana <- diana(population15.data, stand = T) 
plot(cluster.diana, which = 2, main = "", xlab = "", ylab = "") 
```

```r
#区切りを入力
cluster.h <- hclust(dist(population15.data), method = "ward.D2")
plot(cluster.h, hang=-1) 
rect.hclust(cluster.h, k = 3, border = "red") 
```

## 5-4. クラスタ分析：分割型手法(k-means法)

```r
cluster.kmeans = kmeans(population45.data, 6)
cluster.kmeans$cluster
cluster.kmeans$size
cluster.kmeans$withinss
```

```r
clusplot(population45.data, cluster.kmeans$cluster, color=TRUE, shade=TRUE, labels=2, lines=0)
```
